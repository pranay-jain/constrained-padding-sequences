## Figure 2 Plots

This notebook will generate the plots for Figure 2 in the paper. These experiments compare the PFS algorithm to the BDK baseline on the mutual information of each of their produced padding schemes.

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
import os 

# Set the working directory to the parent directory to load the model code
# WARNING: Do NOT run this code block multiple times as it will mutate the working directory.
sys.path.insert(0, '..')
os.chdir('..')
print(os.getcwd())

%load_ext autoreload
%autoreload 2
```

```python
from bdk import run_bdk
from pfs import run_pfs
```

```python
dataset = "autocomplete"

_, backes_min, backes_max, backes_c_list, _, _ = run_bdk(dataset)
pfs_res = run_pfs(dataset, 1.05)
pfs_pad_scheme, pfs_i_inf = pfs_res['pad_scheme'], pfs_res['i_inf']
```

```python
fig, axs = plt.subplots(1, 2, gridspec_kw={'width_ratios': [4,1]}, figsize=(6.4, 5.2))

xvals = np.arange(1,len(pfs_i_inf)+1,1)
axs[0].bar(xvals + 0.2, pfs_i_inf, label="PFS (c = 1.05)", color="lightgrey", width=.4, edgecolor='black')

yerrs = np.array(backes_max) - np.array(backes_min)
axs[0].bar(xvals - 0.2, backes_min, label="BDK", color="black", width=.4, edgecolor='black')
plotline, caplines, barlinecols = axs[0].errorbar(xvals - 0.2, backes_min, yerr=yerrs, fmt="none", color="black", lolims=True)
caplines[0].set_marker('_')
caplines[0].set_markersize(6)


legend = axs[0].legend(loc='lower center', fontsize=14, framealpha=1, handletextpad=0.5, handleheight=0.7, handlelength=0.7, borderpad=0.35)
legend._legend_box.width = 180
axs[0].set_xlim(0,15)
axs[0].set_xticks(np.arange(1,15,1))
axs[0].set_ylim(0,9)
axs[0].set_yticks(np.arange(0,10,1))
#axs[0].grid()
axs[0].set_xlabel('Sequence Length', fontsize=20)
axs[0].set_ylabel('Mutual Information (bits)', fontsize=20)

axs[1].set_xlabel('Method', fontsize=20)
axs[1].set_ylabel('Pad Factor', fontsize=24)
axs[1].boxplot([backes_c_list,[]], whis=[0,100])
axs[1].set_xticklabels(['BDK', 'PFS'], fontsize=14)
axs[1].scatter(2,1.05, color='black')

pad_factor_pos = axs[1].get_position()
pad_factor_pos.x0 = pad_factor_pos.x0 + 0.05
pad_factor_pos.x1 = pad_factor_pos.x1 + 0.05
axs[1].set_position(pad_factor_pos)

plt.show()
```
